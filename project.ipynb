```python
import pandas as pd

# Read the Netflix dataset
data = pd.read_csv("netflix_titles.csv")

# Show the first few rows
data.head(20)
data['title'].unique()


```

```python
data.rename(columns={'show_id':'showid','cast':'Cast'})
```

```python
data.columns=data.columns.str.lower()
data.columns
data.columns.str.replace(' ','_')
data.head(5)
```

```python
 # Check which columns have null values
data.isnull().sum()
(data.isnull().sum() / len(data)) * 100

```

```python
data.dropna(inplace=True)
data
```

```python
data['director'] = data['director'].fillna('Unknown')
data['cast'] = data['cast'].fillna('Unknown')
data['country'] = data['country'].fillna('Unknown')
data['date_added'] = data['date_added'].fillna('Not Available')
data
```

```python
# Remove leading/trailing spaces
data['title'] = data['title'].str.strip()
data['director'] = data['director'].str.strip()

# Convert all text to lowercase (optional)
data['type'] = data['type'].str.lower()
data['country'] = data['country'].str.lower()
data
```

```python
# Check duplicates
data.duplicated().sum()
```

```python
# Remove them
data.drop_duplicates(inplace=True)
```

```python
# Convert date_added to datetime format
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')

```

```python
# Extract useful features like year and month
data['year_added'] = data['date_added'].dt.year
data['month_added'] = data['date_added'].dt.month
```

```python
create_table_query = """
CREATE TABLE IF NOT EXISTS netflix_titles (
    show_id VARCHAR(10),
    type VARCHAR(10),
    title VARCHAR(255),
    director VARCHAR(255),
    cast TEXT,
    country VARCHAR(100),
    date_added VARCHAR(50),
    release_year INT,
    rating VARCHAR(20),
    duration VARCHAR(50),
    listed_in TEXT,
    description TEXT
)
"""
cursor.execute(create_table_query)
connection.commit()
```

```python
for _, row in data.iterrows():
    sql = """
    INSERT INTO netflix_titles 
    (show_id, type, title, director, cast, country, date_added, release_year, rating, duration, listed_in, description)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    # Convert row to a list and select only the needed columns
    # Make sure the order matches your SQL statement columns
    values = (
        row['show_id'], row['type'], row['title'], row['director'], 
        row['cast'], row['country'], row['date_added'], row['release_year'],
        row['rating'], row['duration'], row['listed_in'], row['description']
    )
    cursor.execute(sql, values)
connection.commit()
print("✅ Data inserted successfully into MySQL!")
```
